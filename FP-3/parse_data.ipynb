# parse_data.ipynb

This notebook parses the data files used for the FP-2 assignment. 

<br>
<br>

First let's read the attached data file:

```python
import pandas as pd

df0 = pd.read_csv("owid-co2-data.csv")

df0.describe()
```

<br>
<br>

The dependent and independent variables variables (DVs and IVs) that we are interested in are:

**DVs**:
- CO2 emissions  ("co2" column in the CSV file - annual CO2 emissions per country, measured in metric tons)

**IVs**:
- GDP per capita ("gdp" column in the CSV file)
- Population ("population" column in the CSV file)
- Year ("year" column in the CSV file)


<br>
<br>

Let's extract the relevant columns:

```python

df = df0[  ['co2', 'gdp', 'population', 'year']  ]

df.describe()
```

<br>
<br>

Next let's use the `rename` function to give the columns simpler variable names:

```python
df = df.rename( columns={'co2':'co2_emissions', 'gdp':'gdp_per_capita'} )

df.describe()
```
