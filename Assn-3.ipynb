### Figure 2

Use `data-fig2.npy` for this figure.

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt

a = np.load('data-fig2.npy')

plt.figure()
plt.plot(a[:,0], color = 'red', label = 'Red')
plt.plot(a[:,1], color = 'green', label = 'Green')
plt.plot(a[:,2], color = 'blue', label = 'Blue')
plt.legend(loc='upper left')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

### Figure 3

Use `data-fig3.npy` for this figure.

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt

a = np.load('data-fig3.npy')

plt.figure()
plt.scatter(a[:,0], a[:,1], s=200, color='dodgerblue', alpha=0.7, edgecolors='dodgerblue', linewidth=1)
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

### Figure 4

Use `data-fig4.npy` for this figure.

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt

a = np.load('data-fig5.npy')

plt.figure()
plt.imshow(a, cmap = 'RdBu', origin = 'lower', extent = [0, 40, 0, 120],aspect='auto', vmin=-1, vmax=1)
plt.colorbar()
plt.xticks(np.arange(0, 41, 5))
plt.yticks(np.arange(0, 121, 20))
plt.show()
```
